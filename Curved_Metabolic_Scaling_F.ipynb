```python
# %% [markdown]
# ============================================
# Mammal BMR modeling (Kleiber/MTE vs Kernel-OLS)
# Colab-ready end-to-end pipeline
# ============================================

# %% Imports
import io, os, re, math, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import shapiro
from scipy.optimize import curve_fit
from sklearn.model_selection import KFold

plt.rcParams['figure.dpi'] = 120

# Constants
k_B = 8.617e-5  # eV/K (MTE convention)

# Output dir
OUTDIR = "/content/metabolism_outputs"
os.makedirs(OUTDIR, exist_ok=True)

# %% Helpers
def parse_float(x):
    """Tolerant numeric parser for strings with commas, weird bytes, dashes."""
    try:
        s = str(x)
        s = s.replace('\xa0',' ').replace('',' ').replace(',', '').strip()
        if s in ["", "nan", "None", "NA", "", "-", "", "", "", ".", ""]:
            return np.nan
        return float(s)
    except:
        return np.nan

def looks_like_species(s):
    if not isinstance(s, str): return False
    s = s.replace('', ' ').strip()
    return bool(re.match(r'^[A-Z][a-z]+(?:\s+[a-z][a-z\-\.]+){1,3}$', s))

def pick_column(df, candidates_regex, numeric=False):
    """Pick the first column whose name matches any regex in candidates_regex."""
    cols = list(df.columns)
    for pat in candidates_regex:
        r = re.compile(pat, flags=re.I)
        for c in cols:
            if r.search(str(c)):
                if not numeric: return c
                # if numeric requested, ensure column can be parsed to numeric
                try:
                    test = pd.to_numeric(df[c], errors='coerce')
                    if test.notna().sum() > 0:
                        return c
                except:
                    pass
    return None

def original_r2(y, yhat):
    y = np.asarray(y, float); yhat = np.asarray(yhat, float)
    sse = np.sum((y - yhat)**2)
    sst = np.sum((y - y.mean())**2)
    return 1.0 - sse/sst

def calib_fit(y, yhat):
    X = sm.add_constant(yhat)
    mod = sm.OLS(y, X).fit()
    return float(mod.params[0]), float(mod.params[1])  # intercept, slope

def model_mass_power(M, a, b):
    return a * (M ** b)

def model_MTE(M, TbC, a, b, Ea):
    TbK = TbC + 273.15
    return a * (M ** b) * np.exp(-Ea * (1.0 / (k_B * TbK)))

def compute_metrics(y, yhat, k_params):
    y = np.asarray(y, float); yhat = np.asarray(yhat, float)
    resid = y - yhat
    rmse = float(np.sqrt(np.mean(resid**2)))
    mae  = float(np.mean(np.abs(resid)))
    medae = float(np.median(np.abs(resid)))
    mape = float(np.mean(np.abs(resid) / np.maximum(1e-12, y))) * 100.0
    R2   = original_r2(y, yhat)

    n = len(y); p = k_params
    sigma2 = np.mean(resid**2)
    loglik = -(0.5*n)*(math.log(2*math.pi*sigma2) + 1)
    AIC = 2*p - 2*loglik
    BIC = p*math.log(n) - 2*loglik

    c0, c1 = calib_fit(y, yhat)

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        try:
            sh_p = float(shapiro(resid)[1]) if 3 <= n <= 5000 else np.nan
        except Exception:
            sh_p = np.nan

    try:
        Xbp = sm.add_constant(yhat)
        bp_stat, bp_p, _, _ = het_breuschpagan(resid, Xbp)
        bp_p = float(bp_p)
    except Exception:
        bp_p = np.nan

    return {
        "RMSE": rmse, "MAE": mae, "MedAE": medae, "MAPE_%": mape,
        "R2": R2, "AIC": float(AIC), "BIC": float(BIC),
        "Calib_intercept": c0, "Calib_slope": c1,
        "Shapiro_p": sh_p, "BreuschPagan_p": bp_p
    }

def kfold_cv(df, predictor_builder, target_col="BMR_mlO2_h", k=10, seed=42):
    idx = np.arange(len(df))
    kf = KFold(n_splits=k, shuffle=True, random_state=seed)
    rmses = []; maes = []; r2s = []
    for tr, te in kf.split(idx):
        trdf = df.iloc[tr]; tedf = df.iloc[te]
        pred_func = predictor_builder(trdf)
        y_true = tedf[target_col].to_numpy(dtype=float)
        y_pred = pred_func(tedf)
        rmses.append(np.sqrt(np.mean((y_true - y_pred)**2)))
        maes.append(np.mean(np.abs(y_true - y_pred)))
        r2s.append(original_r2(y_true, y_pred))
    return float(np.mean(rmses)), float(np.std(rmses)), float(np.mean(maes)), float(np.std(maes)), float(np.mean(r2s)), float(np.std(r2s))

# %% Upload
print("Upload the SI HTML (e.g., '6428table1.html') or a CSV with columns like M_g, Tb_C, BMR_mlO2_h.")
uploaded = files.upload()
assert len(uploaded) >= 1, "No file uploaded."

fname = list(uploaded.keys())[0]
raw_bytes = uploaded[fname]
print("Received:", fname, f"({len(raw_bytes)} bytes)")

# %% Parse
def load_and_parse(fname, raw_bytes):
    name = fname.lower()
    if name.endswith((".html", ".htm")):
        # Parse first table then heuristically reduce to species rows
        dfs = pd.read_html(io.BytesIO(raw_bytes))
        if len(dfs) == 0:
            raise ValueError("No tables found in HTML.")
        t = dfs[0].copy()
        t.columns = [str(c) for c in t.columns]
        # Try to find blocks with a species-looking second/third column
        # Standardize column placeholders
        while t.shape[1] < 6:
            t[f"extra{t.shape[1]}"] = np.nan
        # heuristic: species likely in any col with 'Genus species'
        mask_any = pd.Series(False, index=t.index)
        for c in t.columns:
            mask_any = mask_any | t[c].astype(str).apply(looks_like_species)
        sp = t.loc[mask_any].copy()

        # Build a clean DF of candidate columns
        # Try to pick mass, Tb, BMR columns by header names or ranges
        # 1) rename headers to simpler tokens
        ren = {c:c for c in t.columns}
        df = sp.copy()

        # Try name-based selection
        mass_col = pick_column(df, [r'^\s*M\s*[,/ ]*g\s*$', r'\bM\b', r'mass', r'body\s*mass'], numeric=True)
        tb_col   = pick_column(df, [r'^\s*Tb', r'\btemp', r'temperature', r'\bT\b'], numeric=True)
        bmr_col  = pick_column(df, [r'BMR', r'basal.*metab', r'oxygen', r'ml.*O2.*h', r'VO2'], numeric=True)

        # If those fail, brute-force: pick numeric columns by plausible ranges
        cand_numeric = []
        for c in df.columns:
            x = pd.to_numeric(df[c].apply(parse_float), errors='coerce')
            if x.notna().sum() > 0:
                cand_numeric.append((c, x))

        if mass_col is None:
            # mass usually very broad (grams ~ 2..1e6); choose column with largest range
            rngs = [(c, x.max()-x.min()) for c,x in cand_numeric]
            mass_col = max(rngs, key=lambda z: z[1])[0]

        if bmr_col is None:
            # BMR should be positive and roughly 5..1e5 ml/h; choose the non-mass numeric with next largest range
            # crude but works well here
            non_mass = [(c, x) for c,x in cand_numeric if c != mass_col]
            rngs = [(c, x.max()-x.min()) for c,x in non_mass]
            bmr_col = max(rngs, key=lambda z: z[1])[0]

        # Tb: pick numeric col with many values in [25,45]
        if tb_col is None:
            best, bestscore = None, -1
            for c,x in cand_numeric:
                vals = x.dropna()
                if len(vals)==0: continue
                score = ((vals>=25) & (vals<=45)).mean()
                if score > bestscore and c not in (mass_col, bmr_col):
                    best, bestscore = c, score
            tb_col = best  # may remain None (then well just do mass-only)

        # Species and higher taxon (family/order) columns
        species_col = None
        for c in df.columns:
            if df[c].astype(str).apply(looks_like_species).any():
                species_col = c; break

        # A best-effort group column near species (often family/order)
        group_col = None
        if species_col is not None:
            # try preceding column or any string-ish col
            idx = list(df.columns).index(species_col)
            if idx > 0: group_col = df.columns[idx-1]
        if group_col is None:
            group_col = pick_column(df, [r'Order', r'Family', r'group', r'clade', r'Genus'], numeric=False)

        parsed = pd.DataFrame({
            "Family_or_Order": (df[group_col].astype(str).str.replace('',' ', regex=False).str.strip()
                                if group_col is not None else ""),
            "Species": (df[species_col].astype(str).str.replace('',' ', regex=False).str.strip()
                        if species_col is not None else ""),
            "M_g": pd.to_numeric(df[mass_col].apply(parse_float), errors='coerce'),
            "BMR_mlO2_h": pd.to_numeric(df[bmr_col].apply(parse_float), errors='coerce')
        })
        if tb_col is not None:
            parsed["Tb_C"] = pd.to_numeric(df[tb_col].apply(parse_float), errors='coerce')
        else:
            parsed["Tb_C"] = np.nan

        parsed = parsed.dropna(subset=["M_g","BMR_mlO2_h"]).reset_index(drop=True)
        return parsed

    else:
        # CSV
        df = pd.read_csv(io.BytesIO(raw_bytes))
        # Try to map columns by regex
        mass_col = pick_column(df, [r'^\s*M\s*[,/ ]*g\s*$', r'\bM_g\b', r'\bmass', r'body\s*mass', r'\bM\b'], numeric=True)
        bmr_col  = pick_column(df, [r'BMR', r'basal.*metab', r'oxygen', r'ml.*O2.*h', r'VO2'], numeric=True)
        tb_col   = pick_column(df, [r'^\s*Tb', r'\btemp', r'temperature', r'\bT\b'], numeric=True)
        species_col = pick_column(df, [r'species', r'sci.*name', r'Genus.*species'], numeric=False)
        group_col   = pick_column(df, [r'Order', r'Family', r'group', r'clade', r'Genus'], numeric=False)

        if mass_col is None or bmr_col is None:
            raise ValueError("Could not identify M_g and BMR columns in your CSV. "
                             "Expected headers like M_g / mass and BMR_mlO2_h / BMR.")

        parsed = pd.DataFrame({
            "Family_or_Order": (df[group_col].astype(str).str.strip() if group_col else ""),
            "Species": (df[species_col].astype(str).str.strip() if species_col else ""),
            "M_g": pd.to_numeric(df[mass_col].apply(parse_float), errors='coerce'),
            "BMR_mlO2_h": pd.to_numeric(df[bmr_col].apply(parse_float), errors='coerce'),
            "Tb_C": pd.to_numeric(df[tb_col].apply(parse_float), errors='coerce') if tb_col else np.nan
        })
        parsed = parsed.dropna(subset=["M_g","BMR_mlO2_h"]).reset_index(drop=True)
        return parsed

df = load_and_parse(fname, raw_bytes)
print("Parsed rows:", len(df))
display(df.head(10))

# Save the parsed subset for reference
df.to_csv(os.path.join(OUTDIR, "parsed_input.csv"), index=False)

# %% Fit models
# -------- Mass-only dataset
df_mass = df.dropna(subset=["M_g", "BMR_mlO2_h"]).copy()
y_mass = df_mass["BMR_mlO2_h"].to_numpy(float)
M = df_mass["M_g"].to_numpy(float)

# Kernel-OLS (Mass): BMR = c0 + a*M + b*M^2 + d*logM
X_km = pd.DataFrame({
    "const": 1.0,
    "M": M,
    "M_sq": M**2,
    "logM": np.log(M)
})
km_model = sm.OLS(y_mass, X_km).fit()
yhat_km = km_model.predict(X_km)

# Canonical Kleiber (log-OLS back-transform): log(BMR)=loga+b*logM
X_log = sm.add_constant(np.log(M))
cl_model = sm.OLS(np.log(y_mass), X_log).fit()
a_log = math.exp(cl_model.params[0]); b_log = cl_model.params[1]
yhat_cl = a_log * (M ** b_log)

# Canonical Kleiber NLS on raw scale
popt_nls_mass, pcov_nls_mass = curve_fit(model_mass_power, M, y_mass, p0=[a_log, b_log], maxfev=20000)
a_nls, b_nls = popt_nls_mass
yhat_nls_mass = model_mass_power(M, a_nls, b_nls)

# -------- Mass + Tb subset
df_MT = df.dropna(subset=["M_g", "BMR_mlO2_h", "Tb_C"]).copy()
y_MT = df_MT["BMR_mlO2_h"].to_numpy(float)
M_T = df_MT["M_g"].to_numpy(float)
TbC = df_MT["Tb_C"].to_numpy(float)

# Kernel-OLS (Mass+Tb): BMR = c0 + a*M + b*sqrtM + c*Tb + d*M^2 + e*Tb^2
X_kmt = pd.DataFrame({
    "const": 1.0,
    "M": M_T,
    "sqrtM": np.sqrt(M_T),
    "Tb": TbC,
    "M_sq": M_T**2,
    "Tb_sq": TbC**2
})
kmt_model = sm.OLS(y_MT, X_kmt).fit()
yhat_kmt = kmt_model.predict(X_kmt)

# Canonical MTE (log-OLS back-transform): log(BMR)=loga+b*logM - Ea/(kT)
X_mte = pd.DataFrame({"const":1.0, "logM": np.log(M_T), "inv_kT": 1.0 / (k_B*(TbC+273.15))})
clm_model = sm.OLS(np.log(y_MT), X_mte).fit()
loga_mte = clm_model.params["const"]; b_mte = clm_model.params["logM"]; Ea_mte = -clm_model.params["inv_kT"]
a_mte = math.exp(loga_mte)
yhat_clm = a_mte * (M_T ** b_mte) * np.exp(-Ea_mte * (1.0/(k_B*(TbC+273.15))))

# Canonical MTE NLS on raw scale
def f_wrapped(xdata, a, b, Ea):
    Mx, Tbx = xdata
    return model_MTE(Mx, Tbx, a, b, Ea)
popt_mte, pcov_mte = curve_fit(f_wrapped, (M_T, TbC), y_MT, p0=[a_mte, b_mte, Ea_mte], maxfev=30000)
a_nls_m, b_nls_m, Ea_nls_m = popt_mte
yhat_nls_mte = model_MTE(M_T, TbC, a_nls_m, b_nls_m, Ea_nls_m)

# %% Metrics + CV
def predfunc_kernel_mass(train_df):
    M = train_df["M_g"].to_numpy(float)
    y = train_df["BMR_mlO2_h"].to_numpy(float)
    X = pd.DataFrame({"const":1.0,"M":M,"M_sq":M**2,"logM":np.log(M)})
    mod = sm.OLS(y, X).fit()
    def pf(test_df):
        Mt = test_df["M_g"].to_numpy(float)
        Xt = pd.DataFrame({"const":1.0,"M":Mt,"M_sq":Mt**2,"logM":np.log(Mt)})
        return mod.predict(Xt)
    return pf

def predfunc_canon_log_mass(train_df):
    M = train_df["M_g"].to_numpy(float)
    y = train_df["BMR_mlO2_h"].to_numpy(float)
    X = sm.add_constant(np.log(M))
    mod = sm.OLS(np.log(y), X).fit()
    a = math.exp(mod.params[0]); b = mod.params[1]
    def pf(test_df):
        Mt = test_df["M_g"].to_numpy(float)
        return a * (Mt ** b)
    return pf

def predfunc_kernel_MT(train_df):
    M = train_df["M_g"].to_numpy(float)
    Tb = train_df["Tb_C"].to_numpy(float)
    y = train_df["BMR_mlO2_h"].to_numpy(float)
    X = pd.DataFrame({"const":1.0,"M":M,"sqrtM":np.sqrt(M),"Tb":Tb,"M_sq":M**2,"Tb_sq":Tb**2})
    mod = sm.OLS(y, X).fit()
    def pf(test_df):
        Mt = test_df["M_g"].to_numpy(float)
        Tt = test_df["Tb_C"].to_numpy(float)
        Xt = pd.DataFrame({"const":1.0,"M":Mt,"sqrtM":np.sqrt(Mt),"Tb":Tt,"M_sq":Mt**2,"Tb_sq":Tt**2})
        return mod.predict(Xt)
    return pf

def predfunc_canon_log_MTE(train_df):
    M = train_df["M_g"].to_numpy(float)
    Tb = train_df["Tb_C"].to_numpy(float)
    y = train_df["BMR_mlO2_h"].to_numpy(float)
    X = pd.DataFrame({"const":1.0, "logM":np.log(M), "inv_kT":1.0/(k_B*(Tb+273.15))})
    mod = sm.OLS(np.log(y), X).fit()
    a = math.exp(mod.params["const"]); b = mod.params["logM"]; Ea = -mod.params["inv_kT"]
    def pf(test_df):
        Mt = test_df["M_g"].to_numpy(float)
        Tt = test_df["Tb_C"].to_numpy(float)
        return a * (Mt ** b) * np.exp(-Ea * (1.0/(k_B*(Tt+273.15))))
    return pf

# Metrics on full samples
km_metrics   = compute_metrics(y_mass, yhat_km,  int(km_model.df_model)+1)
cl_metrics   = compute_metrics(y_mass, yhat_cl,  2)
nlsm_metrics = compute_metrics(y_mass, yhat_nls_mass, 2)

kmt_metrics   = compute_metrics(y_MT, yhat_kmt, int(kmt_model.df_model)+1)
clm_metrics   = compute_metrics(y_MT, yhat_clm, 3)
nlsM_metrics  = compute_metrics(y_MT, yhat_nls_mte, 3)

# 10-fold CV
cv_mass_ker = kfold_cv(df_mass, predfunc_kernel_mass, k=10)
cv_mass_can = kfold_cv(df_mass, predfunc_canon_log_mass, k=10)
cv_MT_ker   = kfold_cv(df_MT,   predfunc_kernel_MT,   k=10)
cv_MT_can   = kfold_cv(df_MT,   predfunc_canon_log_MTE, k=10)

def row(name, n, p, met, cv=None):
    r = {"Model":name, "n":n, "p":p, **met}
    if cv:
        rm,rs, mm,ms, r2m,r2s = cv
        r.update({"CV_RMSE_mean":rm,"CV_RMSE_sd":rs,
                  "CV_MAE_mean":mm,"CV_MAE_sd":ms,
                  "CV_R2_mean":r2m,"CV_R2_sd":r2s})
    return r

summary_rows = [
    row("Kernel-OLS (Mass)", len(df_mass), int(km_model.df_model)+1, km_metrics, cv=cv_mass_ker),
    row("Canonical Kleiber (log OLS back-transf)", len(df_mass), 2, cl_metrics, cv=cv_mass_can),
    row("Canonical Kleiber (NLS raw)", len(df_mass), 2, nlsm_metrics),
]

if len(df_MT) > 0:
    summary_rows += [
        row("Kernel-OLS (Mass+Tb)", len(df_MT), int(kmt_model.df_model)+1, kmt_metrics, cv=cv_MT_ker),
        row("Canonical MTE (log OLS back-transf)", len(df_MT), 3, clm_metrics, cv=cv_MT_can),
        row("Canonical MTE (NLS raw)", len(df_MT), 3, nlsM_metrics),
    ]

summary_df = pd.DataFrame(summary_rows).round(5)
display(summary_df)
summary_df.to_csv(os.path.join(OUTDIR, "model_comparison_summary.csv"), index=False)

# Coefficient tables (for OLS models)
def coef_table(name, model):
    params = model.params
    if hasattr(params, "index"):
        names = list(params.index)
        coefs = params.values
        bse = model.bse.values if hasattr(model.bse, "values") else np.asarray(model.bse)
        tvals = model.tvalues.values if hasattr(model.tvalues, "values") else np.asarray(model.tvalues)
        pvals = model.pvalues.values if hasattr(model.pvalues, "values") else np.asarray(model.pvalues)
    else:
        names = list(getattr(model.model, "exog_names", [f"p{i}" for i in range(len(params))]))
        coefs = np.asarray(params)
        bse = np.asarray(model.bse)
        tvals = np.asarray(model.tvalues)
        pvals = np.asarray(model.pvalues)
    return pd.DataFrame({"Model":name,"param":names,"coef":coefs,"SE":bse,"t":tvals,"p":pvals})

coef_parts = [coef_table("Kernel-OLS (Mass)", km_model),
              coef_table("Canonical Kleiber (log OLS)", cl_model)]
if len(df_MT)>0:
    coef_parts += [coef_table("Kernel-OLS (Mass+Tb)", kmt_model),
                   coef_table("Canonical MTE (log OLS)", clm_model)]
coef_df = pd.concat(coef_parts, ignore_index=True).round(6)
display(coef_df)
coef_df.to_csv(os.path.join(OUTDIR, "model_coefficients.csv"), index=False)

# %% Local scaling exponent for Kernel-OLS (Mass)
# b_local(M) = (M / yhat) * d yhat / dM, with yhat = c0 + a*M + b*M^2 + d*logM
a = km_model.params.get("M", 0.0)
b = km_model.params.get("M_sq", 0.0)
d = km_model.params.get("logM", 0.0)
c0 = km_model.params.get("const", 0.0)

# Evaluate on a smooth mass grid
M_grid = np.geomspace(max(1e-2, np.nanmin(M)), np.nanmax(M), 300)
yhat_grid = c0 + a*M_grid + b*(M_grid**2) + d*np.log(M_grid)
dy_dM = a + 2*b*M_grid + d*(1.0/M_grid)
b_local = (M_grid / yhat_grid) * dy_dM

# %% Plots
# 1) Observed vs Predicted (Mass-only models)
plt.figure()
plt.scatter(y_mass, yhat_km, s=14, alpha=0.7, label='Kernel-OLS (Mass)')
plt.scatter(y_mass, yhat_cl, s=14, alpha=0.7, label='Kleiber log-OLS->raw')
plt.scatter(y_mass, yhat_nls_mass, s=14, alpha=0.7, label='Kleiber NLS raw')
lims = [0, max(y_mass.max(), yhat_km.max(), yhat_cl.max(), yhat_nls_mass.max())]
plt.plot(lims, lims, lw=1)
plt.xlabel("Observed BMR (ml O2/h)")
plt.ylabel("Predicted BMR (ml O2/h)")
plt.title("Observed vs Predicted (Mass-only)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "obs_vs_pred_mass.png"))
plt.show()

# 2) Residuals vs logM
plt.figure()
plt.scatter(np.log(M), y_mass - yhat_km, s=12, alpha=0.7, label='Kernel-OLS (Mass)')
plt.scatter(np.log(M), y_mass - yhat_cl, s=12, alpha=0.7, label='Kleiber log-OLS->raw')
plt.scatter(np.log(M), y_mass - yhat_nls_mass, s=12, alpha=0.7, label='Kleiber NLS raw')
plt.axhline(0, lw=1)
plt.xlabel("log(Mass [g])")
plt.ylabel("Residual (Observed - Predicted)")
plt.title("Residuals vs log(M)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "residuals_vs_logM.png"))
plt.show()

# 3) Local exponent vs logM (Kernel-OLS mass)
plt.figure()
plt.plot(np.log(M_grid), b_local, lw=2)
plt.xlabel("log(Mass [g])")
plt.ylabel("Local scaling exponent b_local(M)")
plt.title("Local exponent from Kernel-OLS (Mass)")
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "local_exponent_kernel_mass.png"))
plt.show()

# 4) If Tb present: Observed vs Predicted (M+Tb models)
if len(df_MT) > 0:
    plt.figure()
    plt.scatter(y_MT, yhat_kmt, s=14, alpha=0.7, label='Kernel-OLS (Mass+Tb)')
    plt.scatter(y_MT, yhat_clm, s=14, alpha=0.7, label='MTE log-OLS->raw')
    plt.scatter(y_MT, yhat_nls_mte, s=14, alpha=0.7, label='MTE NLS raw')
    lims = [0, max(y_MT.max(), yhat_kmt.max(), yhat_clm.max(), yhat_nls_mte.max())]
    plt.plot(lims, lims, lw=1)
    plt.xlabel("Observed BMR (ml O2/h)")
    plt.ylabel("Predicted BMR (ml O2/h)")
    plt.title("Observed vs Predicted (Mass + Tb)")
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(OUTDIR, "obs_vs_pred_mass_tb.png"))
    plt.show()

# %% (Optional) Per-order R^2 if Family_or_Order exists with multiple levels
if "Family_or_Order" in df_mass.columns and df_mass["Family_or_Order"].astype(str).nunique() > 1:
    groups = []
    for g, sub in df_mass.groupby(df_mass["Family_or_Order"].astype(str)):
        if len(sub) >= 6:
            Mm = sub["M_g"].to_numpy(float)
            yy = sub["BMR_mlO2_h"].to_numpy(float)
            # Predict with fitted global models:
            Xsub = pd.DataFrame({"const":1.0,"M":Mm,"M_sq":Mm**2,"logM":np.log(Mm)})
            yk = km_model.predict(Xsub)
            yc = a_log * (Mm ** b_log)
            groups.append((g, original_r2(yy, yk), original_r2(yy, yc), len(sub)))
    if groups:
        order_df = pd.DataFrame(groups, columns=["Group","R2_KernelMass","R2_Kleiber","n"]).sort_values("R2_KernelMass", ascending=False)
        display(order_df.head(20))
        order_df.to_csv(os.path.join(OUTDIR, "per_group_R2_mass.csv"), index=False)

print("\nSaved outputs to:", OUTDIR)
print(" - model_comparison_summary.csv")
print(" - model_coefficients.csv")
print(" - parsed_input.csv")
print(" - plots: obs_vs_pred_mass.png, residuals_vs_logM.png, local_exponent_kernel_mass.png, (and *_mass_tb.png if Tb present)")
```

```python
# %% [markdown]
# ============================================
# Nature/PNAS Add-On: AIC, Likelihood-Ratio Tests, and Bootstrap CI for b_local(M)
# ============================================

import scipy.stats as stats

# --- AIC and Akaike weights ---
def akaike_weights(aic_values):
    """Compute AIC and Akaike weights from a list of AICs."""
    aic_min = np.min(aic_values)
    delta = aic_values - aic_min
    weights = np.exp(-0.5 * delta) / np.sum(np.exp(-0.5 * delta))
    return delta, weights

# Prepare candidate AIC values
aic_labels = []
aic_values = []

# Always include mass-only models
aic_labels += ["Kernel-OLS (Mass)", "Canonical Kleiber (log OLS)", "Canonical Kleiber (NLS raw)"]
aic_values += [km_metrics["AIC"], cl_metrics["AIC"], nlsm_metrics["AIC"]]

# Add Mass+Tb models if available
if len(df_MT) > 0:
    aic_labels += ["Kernel-OLS (Mass+Tb)", "Canonical MTE (log OLS)", "Canonical MTE (NLS raw)"]
    aic_values += [kmt_metrics["AIC"], clm_metrics["AIC"], nlsM_metrics["AIC"]]

aic_values = np.array(aic_values)
delta_aic, weights_aic = akaike_weights(aic_values)

aic_table = pd.DataFrame({
    "Model": aic_labels,
    "AIC": aic_values,
    "AIC": delta_aic,
    "Akaike_weight": weights_aic
}).sort_values("AIC", ascending=True).reset_index(drop=True)

display(aic_table)
aic_table.to_csv(os.path.join(OUTDIR, "akaike_weights_table.csv"), index=False)

print("\nAIC and Akaike weights saved to:", os.path.join(OUTDIR, "akaike_weights_table.csv"))


# --- Likelihood-Ratio Tests (Kernel vs Canonical) ---
def likelihood_ratio_test(ll_full, ll_reduced, df_full, df_reduced):
    """Compute likelihood-ratio test comparing nested models."""
    LR = 2 * (ll_full - ll_reduced)
    df_diff = df_full - df_reduced
    p_value = 1 - stats.chi2.cdf(LR, df=df_diff)
    return LR, df_diff, p_value

lrt_rows = []

# Mass-only comparison
ll_kernel = km_model.llf
ll_canon = cl_model.llf
LR, df_diff, p = likelihood_ratio_test(ll_kernel, ll_canon,
                                       int(km_model.df_model)+1,
                                       int(cl_model.df_model)+1)
lrt_rows.append(["Mass-only", "Kernel-OLS vs Kleiber (log OLS)", LR, df_diff, p])

# M+Tb comparison (if available)
if len(df_MT) > 0:
    ll_kernel = kmt_model.llf
    ll_canon = clm_model.llf
    LR, df_diff, p = likelihood_ratio_test(ll_kernel, ll_canon,
                                           int(kmt_model.df_model)+1,
                                           int(clm_model.df_model)+1)
    lrt_rows.append(["Mass+Tb", "Kernel-OLS vs MTE (log OLS)", LR, df_diff, p])

lrt_df = pd.DataFrame(lrt_rows, columns=["Subset", "Comparison", "LR_statistic", "df_diff", "p_value"])
display(lrt_df)
lrt_df.to_csv(os.path.join(OUTDIR, "likelihood_ratio_tests.csv"), index=False)
print("Likelihood-ratio results saved to:", os.path.join(OUTDIR, "likelihood_ratio_tests.csv"))


# --- Bootstrap CI for local exponent b_local(M) ---
print("\nRunning bootstrap (500 resamples) for b_local(M)...")

n_boot = 500
M_range = np.geomspace(max(1e-2, np.nanmin(M)), np.nanmax(M), 150)
b_local_boot = np.zeros((n_boot, len(M_range)))

for i in range(n_boot):
    boot_idx = np.random.choice(np.arange(len(df_mass)), len(df_mass), replace=True)
    boot_df = df_mass.iloc[boot_idx]

    M_b = boot_df["M_g"].to_numpy(float)
    y_b = boot_df["BMR_mlO2_h"].to_numpy(float)
    X_b = pd.DataFrame({"const":1.0,"M":M_b,"M_sq":M_b**2,"logM":np.log(M_b)})
    mod_b = sm.OLS(y_b, X_b).fit()

    a_b = mod_b.params.get("M",0)
    b_b = mod_b.params.get("M_sq",0)
    d_b = mod_b.params.get("logM",0)
    c0_b = mod_b.params.get("const",0)

    yhat_b = c0_b + a_b*M_range + b_b*M_range**2 + d_b*np.log(M_range)
    dy_b = a_b + 2*b_b*M_range + d_b*(1/M_range)
    b_local_boot[i,:] = (M_range / yhat_b) * dy_b

# Percentile bootstrap CIs
b_mean = np.mean(b_local_boot, axis=0)
b_low = np.percentile(b_local_boot, 2.5, axis=0)
b_high = np.percentile(b_local_boot, 97.5, axis=0)

# Plot with 95% CI
plt.figure()
plt.plot(np.log(M_range), b_mean, label="Mean b_local(M)", color="navy")
plt.fill_between(np.log(M_range), b_low, b_high, color="skyblue", alpha=0.4, label="95% Bootstrap CI")
plt.axhline(0.75, color="gray", ls="--", lw=1, label="Kleiber 3/4 line")
plt.xlabel("log(Mass [g])")
plt.ylabel("Local scaling exponent b_local(M)")
plt.title("Bootstrap 95% CI for local exponent (Kernel-OLS Mass)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "b_local_bootstrap_CI.png"))
plt.show()

print("Saved bootstrap plot to:", os.path.join(OUTDIR, "b_local_bootstrap_CI.png"))

# --- Summary
print("\n Add-on complete:")
print(" - Akaike weights table  akaike_weights_table.csv")
print(" - Likelihood-ratio tests  likelihood_ratio_tests.csv")
print(" - Bootstrap CI plot for local exponent  b_local_bootstrap_CI.png")
```
